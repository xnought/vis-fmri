```python
import numpy as np
import nibabel as nib
```

```python
PATH = "./ds000144_R1.0.0/sub-01/func/sub-01_task-emotionalfaces_run-01_bold.nii.gz"
```

```python
img = nib.load(PATH)
img_data = np.asanyarray(img.dataobj)
img_data.shape
```

Compared to the anat version, the func has a time dimension I presume. And we have a smaller image size. 

I'm not exactly sure which dim means what so let's figure that out 

```python
import matplotlib.pyplot as plt
plt.style.use("dark_background")

print("FIXING LAST 4th DIM, VARYING 3rd")
fig, axes = plt.subplots(1, 10, figsize=(16,5))
for i, ax in enumerate(axes.flatten()):
	ax.imshow(img_data[:,:,i*3,0])
	ax.set_title(f"{i*3}")
	ax.axis('off')
```

```python
print("FIXING LAST 1st DIM, VARYING 3rd")
fig, axes = plt.subplots(1, 10, figsize=(16,5))
for i, ax in enumerate(axes.flatten()):
	ax.imshow(img_data[0,:,i*3,:])
	ax.set_title(f"{i*3}")
	ax.axis('off')
```

```python
print("FIXING LAST 2nd DIM, VARYING 3rd")
fig, axes = plt.subplots(1, 10, figsize=(16,5))
for i, ax in enumerate(axes.flatten()):
	ax.imshow(img_data[:,0,i*3,:])
	ax.set_title(f"{i*3}")
	ax.axis('off')
```

Looks like to me the shape is (x,y,z,t) where t is time. Let's vary time and view x,y

```python
fig, axes = plt.subplots(1, 10, figsize=(16,5))
for i, ax in enumerate(axes.flatten()):
	ax.imshow(img_data[:,:,img_data.shape[2]//4,i*10])
	ax.set_title(f"{i*10}")
	ax.axis('off')
```

```python
fig, axes = plt.subplots(1, 10, figsize=(16,5))
for i, ax in enumerate(axes.flatten()):
	ax.imshow(img_data[img_data.shape[0]//4,:,:,i*10])
	ax.set_title(f"{i*10}")
	ax.axis('off')
```
